# 1.System Setup


```python
Access_Code = "ADMIN99"
Stock_level = 500
```

# 2.Authentication

```python
user_code = input("Enter code: ")
print("1.View Stock Level => enter 1")
print("2.Ship Items       => enter 2")
print("3.Receive Shipment => enter 3")
print("4.Logout           => enter 4")
if user_code != Access_Code:
    print("Access denied")
    print("--------------------------------------")

else:
    print("--------------------------------------")
    while True:
        choice = input("Enter your choice: ")
        if choice == "1":
            print("View Stock level",Stock_level)
            print("--------------------------------------")
        elif choice == "2":
            ship_items = int (input("Enter ship items :"))
            stock_available = Stock_level - ship_items
            print("Stock available :",stock_available)
            print("--------------------------------------")
        elif choice == "3":
            Receive_shipment = int(input("Enter receive shipment: "))
            stock = stock_available + Receive_shipment
            print("Stock is: ",stock)
            new_shipment = int(input("Enter new shipment: "))
            if new_shipment > stock:
                print("Error: Stock Shortage")
                print("--------------------------------------")    
            else:
                print("Stock is delivered")
                print("--------------------------------------")
        elif choice == "4":
            print("Logout")
            break
        else:
            print("Invalid choice")
            print("--------------------------------------")
```
