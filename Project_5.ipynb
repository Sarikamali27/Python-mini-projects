## Function 1 : Verify the Order

```python
items = ["Chai","Burger","Chai"]
print(items)
```

```python
def verify_order(menu):
    for item in items:
        if item not in menu:
             items.remove(item)
    print(items)
verify_order(menu)
```

## Function 2 : Kitchen Slip

```python
items = set(items)
print(items)
```

## Function 3 : Generate bill

```python
total = 0
for item in items:
    total = total + menu[item]
print("Total price: ",total)
```

```python
Tax = total * gst
print("Tax: ",Tax)
```

```python
bill = total + Tax
print("Total bill: ",bill)
```

# Final Execution

```python
menu = {"Samosa":15,
        "Chai":10,
        "Maggi":30,
        "Sandwich":25,
        "Coffee":20
       }
gst = (0.05,)
def verify_order(order):
    return [item for item in order if item in menu]
def print_kitchen_slip(order):
    print(set(order))
def generate_bill(order,menu):
    total = 0
    for item in order:
        total += menu[item]
    tax = total * gst[0]
    return total + tax
offer_menu = {item:price-2 for item,price in menu.items()}
Rohit = ["Chai","Samosa","Chai","Burger","Maggi"]
new_order = verify_order(Rohit)
print(new_order)
print_kitchen_slip(new_order)
bill = generate_bill(new_order,offer_menu)
print("Final bill: ",bill)
```
