# Task 3:-Slang Replacement

```python
file = open("raw_reviews.txt","r")
c = file.read()
file.close()
c = c.replace("gr8","great").replace("btw","by the way").replace("idk","i do not know").replace("lol","funny")
file = open("raw_reviews.txt","w")
file.write(c)
file.close()
print(c)
```

# Task 4:-Sentence Formatting

```python
file = open("raw_reviews.txt","r")
c = file.read()
file.close()
c = c.replace("\n",".\n")+"."
c = c.replace("it is.","it is")
c = c.replace("damaged .","damaged")
c = c.replace("waste of.","waste of")
c = c.replace("life is.","life is")
c = c.replace("okay.","okay")
c = c.capitalize()
file = open("raw_reviews.txt","w") 
file.write(c)
file.close()
print(c)
```

# Task 5:- Duplicate word removal

```python
file = open("raw_reviews.txt","r")
c = file.read()
file.close()
c = c.replace("fast fast fast","fast")
c = c.replace("quickly quickly quickly","quickly")
c = c.replace("okay product okay product okay","okay product")
c = c.replace("product.","")
c = c.replace("product","product.")
file = open("raw_reviews.txt","w")
file.write(c)
print(c)
file.close()
```
