# 2.Dictionaries (Key-Value Pairs,Methods)

1. Create a dictionary to store mobile phone prices.

```python
price = {
    "iPhone" : 50000,
    "Samsung" : 40000,
    "Oppo" : 20000,
    "OnePlus" : 60000
}
print(price)
```

2. During the sale,the price of "iPhone" is reduced to 48000rs.

```python
price["iPhone"] = 48000
print(price)
```

3. Add the price for the newly arrived phone "Pixel" to the dictionary.

```python
price.setdefault("Pixel",30000)
print(price)
```

# 3.Sets (Unique Elements,Set Operations)

1. You are given two lists of customer names.

```python
morning_visitors = {"Sakshi","Pooja","Nita","Samyukta","Anjali"}
evening_visitors = {"Nita","Kalyani","gita","Anjali","Pooja","Sakshi"}
```

2. Some customers visited the store both in the morning and evening.

```python
print(morning_visitors & evening_visitors)
```

3. Find the unique customers who visited the store today.

```python
print(morning_visitors | evening_visitors)
```

# 4.Comprehensions (Flash Sale Discount)

1. Apply Flash Sale discount of 10% on all mobile phones using dictionary comprehensions.

```python
discounted_price = {item: price * 0.9 for item,price in price.items()}
print(discounted_price)
```

2. Create a list containing only the discounted prices using list comprehensions.

```python
price = [48000,40000,20000,60000,30000]
discounted_price = [ price * 0.9 for price in price]
print(discounted_price)
```
